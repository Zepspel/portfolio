```python
def merge(graph, s, t):

    adj_s = graph[s]
    adj_t = graph[t]

    for node, w in adj_t:
        if node == s:                         # если они были связаны, то убираем петлю
            adj_s.remove([t,w])
        else:
            found = next((item for item in adj_s if item[0] == node), None)
            if found:         # если обе s и t соединены с node 
                found[1] += w     # запись про node в листе смежности s                    #  w(st,v)=w(s,v)+w(t,v)
                found_about_s = next((item for item in graph[node] if item[0] == s), None)
                found_about_s[1] += w    # запись про s в листе смежности node
                found_about_t = next((item for item in graph[node] if item[0] == t), None)
                graph[node].remove(found_about_t)
            else:
                adj_s += [[node, w]]
                found_about_t = next((item for item in graph[node] if item[0] == t), None)
                found_about_t[0] = s


    del graph[t]

    return
    
```

```python
def minCutPhase(graph):
    A = set()                  # временное множество выбранных вершин
    a = next(iter(graph))
    A.add(a)

    dist = heapdict.heapdict() # бинарная куча, содержащая дистанции до A
    dist.update({i: 0 for i in graph.keys() if i != a})
    for neigh, w in graph[a]:
        dist[neigh] -= w       # минус так как хотим максимальную кучу

    s = None
    t = a
    
    while len(A) < len(graph):
        
        node, val = dist.popitem()   # достаем ключ исходя из максимального значения
        s, t = t, node
        A.add(node)

        
        for neigh, w in graph[node]:
            if not neigh in A:
                dist[neigh] -= w

    cut_value = -val

    return cut_value, s, t
```
